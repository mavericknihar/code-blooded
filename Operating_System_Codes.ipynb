```python
import threading as thread
import random

# Shared variable
global x
x = 0

# Lock for synchronization
lock = thread.Lock()

# Reader function to read shared data
def Reader():
    print("Reader is Reading")
    lock.acquire()
    print("Shared Data:", x)
    lock.release()
    print()

# Writer function to write to shared data
def Writer():
    global x
    print("Writer is Writing")
    lock.acquire()
    x += 1
    print("Writer is releasing the lock")
    lock.release()
    print()

# Main program
if __name__ == '__main__':
    # Loop to simulate multiple iterations
    for i in range(0, 1):
        # Generate a random number
        randomNumber = random.randint(1, 100)

        # Choose between Reader and Writer based on random number
        if(randomNumber > 50):
            # Create and start a Reader thread
            Thread1 = thread.Thread(target=Reader)
            Thread1.start()
            print(end='\n')
        else:
            # Create and start a Writer thread
            Thread2 = thread.Thread(target=Writer)
            Thread2.start()
            print(end='\n')

        # Wait for threads to finish
        Thread1.join()
        Thread2.join()
```

```python
#Program to implement FCFS(First Come - First Serve) scheduling algorithm

def findWaitingTime(processes, n , bt, wt):
  wt[0] = 0
  for i in range (1 , n):
    wt[i] = bt[i -1] + wt[i -1]
def findTurnAroundTime(processes,n, bt, wt,tat):
      for i in range(n):
        tat[i] = bt[i] + wt[i]
def findavgTime(processes, n, bt):
  wt = [0] * n
  tat = [0] * n
  total_wt = 0
  total_tat = 0
  findWaitingTime(processes, n , bt, wt)
  findTurnAroundTime(processes, n, bt, wt, tat)
  print("Processes Burst time" + "Waiting Time" + "Turn Around Time")
  for i in range(n):
    total_wt = total_wt + wt[i]
    total_tat = total_tat + tat[i]
    print("" + str(i + 1) + "\t\t" + str(bt[i]) + "\t" + str(wt[i]) + "\t\t" + str(tat[i]))
    print("Average Wait time = " + str(total_wt /n))
    print("Average turn around time =" + str(total_tat /n))
if __name__ == "__main__":
  processes = [1,2,3]
  n = len(processes)
  burst_time = [10,5,8]
  findavgTime(processes, n, burst_time)
```

```python
#Program to work with Semaphores
import threading
import time

CAPACITY = 10
buffer = [-1 for i in range(CAPACITY)]
in_index = 0
out_index = 0
mutex = threading.Semaphore(1)  # Initialize mutex with an initial value of 1
empty = threading.Semaphore(CAPACITY)  # Initialize empty to the size of the buffer
full = threading.Semaphore(0)  # Initialize full to 0

class Producer(threading.Thread):
    def run(self):
        global CAPACITY, buffer, in_index
        global mutex, empty, full
        items_produced = 0
        counter = 0
        while items_produced < 10:  # Fix the loop condition
            empty.acquire()
            mutex.acquire()
            counter += 1
            buffer[in_index] = counter
            in_index = (in_index + 1) % CAPACITY
            print("Producer produced:", counter)
            mutex.release()
            full.release()
            time.sleep(0.1)  # Fix the sleep duration
            items_produced += 1

class Consumer(threading.Thread):
    def run(self):
        global CAPACITY, buffer, out_index
        global mutex, empty, full
        items_consumed = 0
        while items_consumed < 10:
            full.acquire()
            mutex.acquire()
            item = buffer[out_index]
            out_index = (out_index + 1) % CAPACITY
            print("Consumer consumed item:", item)
            mutex.release()
            empty.release()
            time.sleep(0.1)
            items_consumed += 1

if __name__ == "__main__":
    producer = Producer()
    consumer = Consumer()
    consumer.start()
    producer.start()
    producer.join()
    consumer.join()
```

```python
#Program to generate Fibonnacci Series in Python using multithreaded programming
import threading

def print_fibboseries():
    nterms = int(input("How many terms do you want?"))
    n1, n2 = 0, 1
    count = 0
    if nterms <= 0:
        print("Please enter a positive integer.")
    elif nterms == 1:
        print("Fibonacci Sequence up to", nterms, ",")
        print(n1)
    else:
        print("Fibonacci Series")
        while count < nterms:
            print(n1)
            nth = n1 + n2
            n1 = n2
            n2 = nth
            count += 1

if __name__ == "__main__":
    t1 = threading.Thread(target=print_fibboseries)
    t1.start()
    t1.join()
    print("Done")
```

```python
import heapq as hq

SIZE = 4

def sjfNonpreemptive(arr):
    # Use meaningful variable names, 'j' is not defined
    index = 0
    for i in range(SIZE - 1):
        index = i
    for i in range(i + 1, SIZE):
        if (arr[i][1] < arr[index][1]):  # Fix the comparison condition
            index = i
    arr[i], arr[index] = arr[index], arr[i]  # Fix the indentation

    ctime = arr[0][1]
    wait = []
    temp = arr[0][1]
    hq.heappush(wait, arr[0].copy())
    arr[0][1] = -1
    print("Process id", end="\t")
    print("Arrival time", end="\t")
    print("Burst time", end="\t")
    print()

    while wait:  # Fix the condition
        print(end="\t")
        print(wait[0][2], end="\t")
        print(wait[0][1], end="\t")
        print(wait[0][0], end="\t")
        print()
        ctime += wait[0][0]
        hq.heappop(wait)

    for i in range(SIZE):
        if (arr[i][1] <= ctime and arr[i][1] != -1):
            hq.heappush(wait, arr[i].copy())
            arr[i][1] = -1

if __name__ == "__main__":
    # Initialize an array of process info structures.
    arr = [None] * SIZE
    arr[0] = [3, 4, "p1"]
    arr[1] = [8, 0, "p2"]
    arr[2] = [4, 5, "p3"]
    arr[3] = [2, 9, "p4"]
    print("Process scheduling according to SJF is: \n")
    sjfNonpreemptive(arr)
```
